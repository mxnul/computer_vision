```python
import os

def print_tree(root, max_files=5):
    for rootdir, dirs, files in os.walk(root):
        level = rootdir.replace(root, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f"{indent}{os.path.basename(rootdir)}/")
        subindent = ' ' * 4 * (level + 1)
        for f in files[:max_files]:
            print(f"{subindent}{f}")
        if len(files) > max_files:
            print(f"{subindent}... ({len(files)} files total)")

print_tree("/content/source_dataset", max_files=20)
```

```python
import shutil
from pathlib import Path

SOURCE_DIR = Path("/content/source_dataset")
NORMALIZED_DIR = Path("/content/normalized_dataset")
ALLOWED_EXTENSIONS = {'.jpg', '.jpeg', '.png', '.bmp', '.gif'}

# Clean or create normalized dir
if NORMALIZED_DIR.exists():
    print("Removing existing normalized dir to ensure clean copy...")
    shutil.rmtree(NORMALIZED_DIR)
NORMALIZED_DIR.mkdir(parents=True, exist_ok=True)


classes_found = []
for cls in SOURCE_DIR.iterdir():
    if cls.is_dir():
        classes_found.append(cls.name)
        target_cls_dir = NORMALIZED_DIR / cls.name
        target_cls_dir.mkdir(parents=True, exist_ok=True)
        count = 0
        for f in cls.rglob("*"):
            if f.is_file() and f.suffix.lower() in ALLOWED_EXTENSIONS:
                # make safe filename: replace spaces and dots with underscore, lowercase
                new_stem = f.stem.replace(" ", "_").replace(".", "_").lower()
                new_name = new_stem + ".jpg"
                dst = target_cls_dir / new_name
                # if collision, append number
                i = 1
                while dst.exists():
                    dst = target_cls_dir / f"{new_stem}_{i}.jpg"
                    i += 1
                shutil.copy2(f, dst)
                count += 1
        print(f"Copied {count} images for class '{cls.name}'")
if not classes_found:
    # fallback: if there are images directly under SOURCE_DIR
    print("No class folders found at top level. Looking for images to put into a single class 'all_images'.")
    target_cls_dir = NORMALIZED_DIR / "all_images"
    target_cls_dir.mkdir(parents=True, exist_ok=True)
    count = 0
    for f in SOURCE_DIR.rglob("*"):
        if f.is_file() and f.suffix.lower() in ALLOWED_EXTENSIONS:
            new_stem = f.stem.replace(" ", "_").replace(".", "_").lower()
            new_name = new_stem + ".jpg"
            dst = target_cls_dir / new_name
            i = 1
            while dst.exists():
                dst = target_cls_dir / f"{new_stem}_{i}.jpg"
                i += 1
            shutil.copy2(f, dst)
            count += 1
    print(f"Copied {count} images into 'all_images' class. You should relabel classes if needed.")
else:
    print("Normalization complete. Normalized dataset lives at /content/normalized_dataset")
```

```python
import random
from pathlib import Path
import shutil

RANDOM_SEED = 123
VAL_SPLIT = 0.15
TEST_SPLIT = 0.15

SRC = Path("/content/normalized_dataset")
WORK_DIR = Path("/content/dataset")
if WORK_DIR.exists():
    print("Removing existing /content/dataset to avoid leftover files...")
    shutil.rmtree(WORK_DIR)
WORK_DIR.mkdir(parents=True, exist_ok=True)

random.seed(RANDOM_SEED)

classes = sorted([p for p in SRC.iterdir() if p.is_dir()])
if not classes:
    raise FileNotFoundError("No class folders found in /content/normalized_dataset. Check previous cell.")

for cls in classes:
    files = list(cls.glob("*"))
    if not files:
        print(f"WARNING: class {cls.name} has 0 images")
        continue
    random.shuffle(files)
    n = len(files)
    n_test = int(n * TEST_SPLIT)
    n_val = int(n * VAL_SPLIT)
    test_files = files[:n_test]
    val_files = files[n_test:n_test+n_val]
    train_files = files[n_test+n_val:]

    for subset_name, subset_files in [("train", train_files), ("val", val_files), ("test", test_files)]:
        out_dir = WORK_DIR / subset_name / cls.name
        out_dir.mkdir(parents=True, exist_ok=True)
        for src_f in subset_files:
            shutil.copy2(src_f, out_dir / src_f.name)
    print(f"Class {cls.name}: total {n}, train {len(train_files)}, val {len(val_files)}, test {len(test_files)}")

print("Splits created in /content/dataset")
```

```python
import tensorflow as tf
from pathlib import Path

AUTOTUNE = tf.data.AUTOTUNE
IMG_SIZE = (128, 128)   # keeps RAM low
BATCH_SIZE = 16         # lower batch size for memory

WORK_DIR = Path("/content/dataset")
RANDOM_SEED = 123

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    WORK_DIR/"train",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=RANDOM_SEED
)
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    WORK_DIR/"val",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)
test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    WORK_DIR/"test",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

class_names = train_ds.class_names
print("Detected classes:", class_names)

# Do not cache full dataset in RAM to avoid OOM; only prefetch
train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.prefetch(buffer_size=AUTOTUNE)
```

```python
from tensorflow.keras import layers, Sequential

data_augmentation = Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1),
])

normalization_layer = layers.Rescaling(1./255)

# Apply augmentation
train_ds_aug = train_ds.map(lambda x, y: (data_augmentation(x, training=True), y))
# Normalize for all sets
train_ds_norm = train_ds_aug.map(lambda x, y: (normalization_layer(x), y)).prefetch(tf.data.AUTOTUNE)
val_ds_norm = val_ds.map(lambda x, y: (normalization_layer(x), y)).prefetch(tf.data.AUTOTUNE)
test_ds_norm = test_ds.map(lambda x, y: (normalization_layer(x), y)).prefetch(tf.data.AUTOTUNE)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Sequential

num_classes = len(class_names)

model_cnn = Sequential([
    layers.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3)),
    layers.Conv2D(32, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(128, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')
])

model_cnn.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

model_cnn.summary()
```

```python
EPOCHS = 7

history_cnn = model_cnn.fit(
    train_ds_norm,
    validation_data=val_ds_norm,
    epochs=EPOCHS
)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, Sequential

base_model = MobileNetV2(input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3),
                         include_top=False,
                         weights='imagenet')
base_model.trainable = False  # freeze base to save memory and speed up

model_mobilenet = Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')
])

model_mobilenet.compile(optimizer='adam',
                        loss='sparse_categorical_crossentropy',
                        metrics=['accuracy'])

model_mobilenet.summary()
```

```python
EPOCHS = 7

history_mobilenet = model_mobilenet.fit(
    train_ds_norm,
    validation_data=val_ds_norm,
    epochs=EPOCHS
)
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

# collect true labels from test set
y_true = np.concatenate([y.numpy() for x, y in test_ds_norm], axis=0)

# Predict function helper
def get_preds(model, dataset):
    preds = model.predict(dataset, verbose=1)
    return np.argmax(preds, axis=1)

y_pred_cnn = get_preds(model_cnn, test_ds_norm)
y_pred_mobilenet = get_preds(model_mobilenet, test_ds_norm)

print("=== CNN Classification Report ===")
print(classification_report(y_true, y_pred_cnn, target_names=class_names))

print("=== MobileNetV2 Classification Report ===")
print(classification_report(y_true, y_pred_mobilenet, target_names=class_names))
```

```python
import matplotlib.pyplot as plt

def plot_history(history, title):
    acc = history.history.get('accuracy', [])
    val_acc = history.history.get('val_accuracy', [])
    loss = history.history.get('loss', [])
    val_loss = history.history.get('val_loss', [])
    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(epochs, acc, 'b-o', label='Train Acc')
    plt.plot(epochs, val_acc, 'r-o', label='Val Acc')
    plt.title(f"{title} - Accuracy")
    plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.grid(True)

    plt.subplot(1,2,2)
    plt.plot(epochs, loss, 'b-o', label='Train Loss')
    plt.plot(epochs, val_loss, 'r-o', label='Val Loss')
    plt.title(f"{title} - Loss")
    plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.grid(True)

    plt.show()

plot_history(history_cnn, "CNN")
plot_history(history_mobilenet, "MobileNetV2")
```
